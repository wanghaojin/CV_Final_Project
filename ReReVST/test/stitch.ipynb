```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

path1 = 'result_frames_1/ReReVST-kandinsky-inputfox1/'
path2 = 'result_frames_2/ReReVST-kandinsky-inputfox2/'
path3 = 'result_frames_3/ReReVST-kandinsky-inputfox3/'
path4 = 'result_frames_4/ReReVST-kandinsky-inputfox4/'
output_path = 'result_balabala'
os.makedirs(output_path,exist_ok=True)
def get_frames(img1,img2,img3,img4):
    res = np.zeros((224,224,3))
    res[0:112,0:112] = img1
    res[0:112,112:224] = img2
    res[112:224,0:112] = img3
    res[112:224,112:224] = img4
    return res

for img in os.listdir(path1):
    name = img.split('.')[0]
    img1 = cv2.imread(os.path.join(path1,img))
    img2 = cv2.imread(os.path.join(path2,name + '.jpg'))
    img3 = cv2.imread(os.path.join(path3,name + '.jpg'))
    img4 = cv2.imread(os.path.join(path4,name + '.jpg'))
    print(img1.shape)
    print(img2.shape)
    print(img3.shape)
    print(img4.shape)
    frames = get_frames(img1,img2,img3,img4)
    cv2.imwrite(os.path.join(output_path,name + '.jpg'),frames)
print('frame' + ' done')
    
```
